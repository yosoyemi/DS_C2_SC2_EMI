```python
import pandas as pd
from pathlib import Path
import kagglehub
import zipfile, io

# ================================================================
# OPCIÓN A – leer archivo local ZIP:  ................................
local_zip = Path("/content/archive (1).zip")   # ajusta el nombre si cambia
if local_zip.exists():
    with zipfile.ZipFile(local_zip, "r") as zf:
        with zf.open("country_vaccinations.csv") as f:
            df = pd.read_csv(f)
else:
    # ==============================================================
    # OPCIÓN B – descargar directo de Kaggle: ........................
    # (requiere tu token de la API configurado en Colab)
    df = kagglehub.load_dataset(
        kagglehub.KaggleDatasetAdapter.PANDAS,
        "gpreda/covid-world-vaccination-progress",
        "country_vaccinations.csv",
    )

print(f"Filas cargadas: {len(df):,}")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d", errors="coerce")

num_cols = [
    "total_vaccinations", "people_vaccinated", "people_fully_vaccinated",
    "daily_vaccinations", "total_vaccinations_per_hundred",
    "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred",
]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

df.info()
```

```python
vac_por_compania = (
    df.assign(vaccine_list=df["vaccines"].str.split(r",\s*"))
      .explode("vaccine_list")
      .groupby("vaccine_list", as_index=False)["daily_vaccinations"]
      .sum()
      .rename(columns={"vaccine_list": "compania",
                       "daily_vaccinations": "dosis_aplicadas"})
      .sort_values("dosis_aplicadas", ascending=False)
)
vac_por_compania.head()
```

```python
total_mundial = df["daily_vaccinations"].sum()
print(f"Total de dosis aplicadas: {total_mundial:,}")
```

```python
promedio_pais = (
    df.groupby("country", as_index=False)["daily_vaccinations"]
      .mean()
      .rename(columns={"daily_vaccinations": "promedio_diario"})
      .sort_values("promedio_diario", ascending=False)
)
promedio_pais.head(10)
```

```python
vac_29ene = df.loc[df["date"] == "2021-01-29", "daily_vaccinations"].sum()
print(f"Dosis aplicadas el 29/01/2021: {vac_29ene:,}")
```

```python
if "daily_vaccinations_raw" not in df.columns:
    df["daily_vaccinations_raw"] = pd.NA

conDiferencias = df.copy().assign(
    diferencias=lambda x: x["daily_vaccinations"] - x["daily_vaccinations_raw"]
)
conDiferencias.head()
```

```python
dias = (df["date"].max() - df["date"].min()).days
print(f"El dataset cubre {dias} días "
      f"({df['date'].min().date()} → {df['date'].max().date()}).")
```

```python
conCantidad = df.copy().assign(
    canVac=lambda x: x["vaccines"].str.split(r",\s*").str.len()
)
conCantidad.head()
```

```python
antes20 = df[df["date"] < "2020-12-20"].copy()
antes20.head()
```

```python
pfizer = df[df["vaccines"].str.contains("Pfizer", case=False, na=False)].copy()
pfizer.head()
```

```python
nombre_excel = "resultadosReto.xlsx"
with pd.ExcelWriter(nombre_excel, engine="openpyxl") as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias", index=False)
    conCantidad.to_excel(writer, sheet_name="conCantidad", index=False)
    antes20.to_excel(writer, sheet_name="antes20", index=False)
    pfizer.to_excel(writer, sheet_name="pfizer", index=False)

print(f"Archivo creado: {nombre_excel}")
```
